```python
import pandas as pd
```

```python
phones = pd.read_csv('Data/smartphones.csv') 
```

```python
phones.head()
```

# 1.Agrupación por Marca y Promedio de Precios Utiliza groupby para agrupar los smartphones por marca y calcula el precio promedio de cada una.



```python
# Agrupar por región y sumar las ventas
phones.groupby('Brand')['Final Price'].sum()
```

# 2. Conteo de Modelos por Marca Emplea groupby para contar la cantidad de modelos diferentes que tiene cada marca.

```python
phones.groupby('Brand')['Model'].sum()
```

# 3. RAM Promedio por Marca Usa groupby para determinar la cantidad promedio de RAM para cada marca de smartphone

```python
phones.groupby('Brand')['RAM'].mean()
```

# 4.Tabla Pivot de Precios Máximos por Color Crea una tabla pivot que muestre el precio máximo de los smartphones para cada color

```python
phones.pivot_table(values='Final Price', index='Color', aggfunc='max')
```

# 5. Distribución de Almacenamiento por Marca Con groupby, calcula la distribución de la capacidad de almacenamiento para cada marca.

```python
phones.groupby('Brand')['Storage'].value_counts()
```

# 6. Precio Promedio por Combinación de RAM y Almacenamiento Genera una tabla pivot que muestre el precio promedio para cada combinación única de RAM y almacenamiento.

```python
phones.pivot_table(values='Final Price', index=['RAM','Storage'], aggfunc='mean')
```

# 7. Cantidad de Smartphones Libres por Marca Utiliza groupby para contar cuántos smartphones de cada marca están libres de contratos con compañías celulares.

```python
phones.groupby('Brand')['Smartphone'].count()
```

# 8. Precio Mínimo, Máximo y Promedio por Marca Crea una tabla pivot que resuma el precio mínimo, máximo y promedio para cada marca de smartphone.

```python
phones.pivot_table(values='Final Price', index=('Brand'), aggfunc=['min','max', 'mean'])
```

# 9. Smartphones por Rango de Precios Emplea groupby para agrupar los smartphones en rangos de precios y cuenta cuántos hay en cada rango.

```python
phones['Range Price'] = pd.cut(phones['Final Price'], bins=[0,300,600,900,1200,1500,1800,1900,2400])
```

```python
phones.groupby('Range Price').size()
```

# 10. Comparación de Precios entre Smartphones Libres y con Contrato Genera una tabla pivot que compare el precio promedio de los smartphones libres (Free = Yes) versus los que están atados a contratos (Free = No).

```python
phones.pivot_table(values='Final Price', index=('Free'), aggfunc='mean')
```
